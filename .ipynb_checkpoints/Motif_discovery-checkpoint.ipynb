```python
# import libraries 
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import seaborn as sns

from matplotlib import patches
from matplotlib_venn import venn2, venn3
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import fdrcorrection
```

# Import results
https://prosite.expasy.org/scanprosite/scanprosite_doc.html#of_graphical


```python
data_path = os.getcwd() + "/Motif_discovery/"
fig_path = os.getcwd() + "/Figures/"
```

## PROSITE full scan

```python
prosite_CSF_pos = pd.read_csv(data_path + "PROSITE_CSF_positive.txt", sep="\t", index_col=False, header=None, 
    names=["Uniprot", "Start", "End", "PROSITE motif", "Profile score", "Confidence (miniprofile)", "Confidence", 
           "Matched region"], skiprows=7)
prosite_CSF_pos_short = prosite_CSF_pos.loc[:, ["Uniprot", "PROSITE motif"]]
prosite_CSF_pos_short.head(n=20)
```

```python
prosite_CSF_neg = pd.read_csv(data_path + "PROSITE_CSF_negative.txt", sep="\t", index_col=False, header=None, 
    names=["Uniprot", "Start", "End", "PROSITE motif", "Profile score", "Confidence (miniprofile)", "Confidence", 
           "Matched region"], skiprows=7)
prosite_CSF_neg_short = prosite_CSF_neg.loc[:, ["Uniprot", "PROSITE motif"]]
prosite_CSF_neg_short.head(n=20)
```

```python
def get_uniprot(string):
    _, uniprot, _ = string.split("|")
    return uniprot
```

```python
# retrieve Uniprot ID
prosite_CSF_pos_short["Uniprot"] = prosite_CSF_pos_short["Uniprot"].apply(get_uniprot)
# retrieve PROSITE ID and pattern name
prosite_CSF_pos_short[["PROSITE motif", "Pattern name"]] = prosite_CSF_pos_short["PROSITE motif"].str.split("|", expand=True)
# drop duplicate pattern in same protein
prosite_CSF_pos_short.drop_duplicates(subset = ["Uniprot", "PROSITE motif"], inplace=True)
prosite_CSF_pos_short
```

```python
# retrieve Uniprot ID
prosite_CSF_neg_short["Uniprot"] = prosite_CSF_neg_short["Uniprot"].apply(get_uniprot)
# retrieve PROSITE ID and pattern name
prosite_CSF_neg_short[["PROSITE motif", "Pattern name"]] = prosite_CSF_neg_short["PROSITE motif"].str.split("|", expand=True)
# drop duplicate pattern in same protein
prosite_CSF_neg_short.drop_duplicates(subset = ["Uniprot", "PROSITE motif"], inplace=True)
prosite_CSF_neg_short
```

```python
pattern_count_CSF_pos = prosite_CSF_pos_short["Pattern name"].value_counts()
pattern_count_CSF_neg = prosite_CSF_neg_short["Pattern name"].value_counts()
```

```python
# overlap of PROSITE patterns
pos_set = set(prosite_CSF_pos_short["PROSITE motif"])
neg_set = set(prosite_CSF_neg_short["PROSITE motif"])

venn2([pos_set, neg_set], set_labels=("CSF positive", "CSF negative"))
plt.show()
```

```python
# merge positive and negative CSF associated patterns
pattern_count = pd.concat([pattern_count_CSF_pos, pattern_count_CSF_neg], axis=1)
pattern_count.fillna(0, inplace=True)
# pattern_count.reset_index(inplace=True)
# pattern_count.columns = ["Pattern", "CSF positive", "CSF negative"]
pattern_count.columns = ["CSF positive", "CSF negative"]
pattern_count
```

```python
def fishers_test(df):

    A = int(df.values[0:1]) # CSF positive & pattern yes
    B = int(num_CSF_pos - df.values[0:1]) # CSF positive & pattern no
    C = int(df.values[1:2]) # CSF negative & pattern yes
    D = int(num_CSF_neg - df.values[1:2]) # CSF negative & pattern no

    # create 2x2 contingency table of counts
    table = np.array([[A, C], [B, D]])
    table
    odds, p = fisher_exact(table, alternative="two-sided") # function return odds ratio and p-value
    return p
```

```python
# relative abundance of patterns in CSF positive and negative protein set
num_CSF_pos = 970
num_CSF_neg = 1530

pattern_count["CSF positive percentage"] = pattern_count["CSF positive"].apply(lambda x: x/num_CSF_pos)
pattern_count["CSF negative percentage"] = pattern_count["CSF negative"].apply(lambda x: x/num_CSF_neg)
pattern_count["Difference"] =  pattern_count["CSF positive"] - pattern_count["CSF negative"]
# calculate p-value (Fisher's exact test, chi square)
pattern_count["P-val"] = pattern_count.apply(fishers_test, axis=1)
# calculate adjusted p-value (FDR correction)
rejected, p_adj = fdrcorrection(pattern_count["P-val"])
pattern_count["P-val adjusted"] = p_adj
pattern_count.sort_values("P-val adjusted", inplace=True)
```

```python
# look only at domain with significant difference between groups
pattern_count_sig = pattern_count[pattern_count["P-val adjusted"] < 0.05]
pattern_count_sig.sort_values("Difference", ascending=False, inplace=True)
pattern_count_sig
```

```python
# create bar plots for every PROSITE pattern and save figure
for pat in pattern_count_sig.index:
#     print(pat)
    # create barplot of percentages in CSF positive and negative
    plt.bar(x=pattern_count_sig.loc[pat].index[2:4], height=pattern_count_sig.loc[pat].values[2:4])
    plt.title(pat)
    # save figure
    plt.gcf()
    plt.savefig(data_path + "PROSITE_patterns/" + pat + ".png")
    plt.clf()
```

```python
# create melted dataframe
pattern_count_sig_less = pattern_count_sig[["CSF positive percentage", "CSF negative percentage"]]
pattern_count_sig_less.reset_index(inplace=True)
pattern_count_melted = pattern_count_sig_less.melt(id_vars=["index"], value_vars=["CSF positive percentage", "CSF negative percentage"], 
    var_name="CSF", value_name="Percentage")
```

```python
# summary figure 
sns.set(style=("ticks"), font_scale=2)
palette = {"CSF positive percentage": "cornflowerblue", "CSF negative percentage": "palevioletred"}

fig = plt.figure(figsize=(18, 12)) 
bar = sns.barplot(x="index", y="Percentage", data=pattern_count_melted, hue="CSF", palette=palette)
bar.legend(title="")
bar.set(xlabel=None, ylabel="Percentage", title="Percentage of brain proteins associated with domains and patterns")

# create legend
P = patches.Patch(color="cornflowerblue", label="Present")
N = patches.Patch(color="palevioletred", label="Not present")
plt.legend(title="CSF", loc="upper center", handles=[P, N]);

fig.autofmt_xdate(rotation=90)
plt.show()

fig.savefig(fig_path + "PROSITE_domains_brain_proteins.png", bbox_inches="tight")
```
