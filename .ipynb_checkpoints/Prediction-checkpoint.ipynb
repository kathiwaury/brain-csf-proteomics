```python
def get_brain_expression(string):
    # check if expression for multiple tissues is provided
    if ";" in string:
        tissues = string.split(";")
        for t in tissues:
            # keep only information on brain expression
            if "brain" in t:
                brain_string = t 
    else:
        brain_string = string
    
    # extract expression value from string
    _, exp = brain_string.split(" ")
    exp = float(exp)

    return exp

# add column with the mRNA brain expression
brain_elevated["Brain expression"] = brain_elevated["RNA tissue specific nTPM"].apply(get_brain_expression)

#### TO DO ####
# extract brain region-specific expressions (limited)
```

# Create feature data set

```python
brain_all = pd.read_csv(os.getcwd() + "/Motif_discovery/Brain_all.txt", header=None, error_bad_lines=False)
CSF_neg = pd.read_csv(os.getcwd() + "/Motif_discovery/Brain_CSF_negative_all.txt", header=None, error_bad_lines=False)
CSF_pos = pd.read_csv(os.getcwd() + "/Motif_discovery/Brain_CSF_positive_all.txt", header=None, error_bad_lines=False)
feature_df = pd.read_csv(os.getcwd() + "/Datasets/features_human_proteome.csv")
feature_df
```

```python
feature_df_brain = feature_df.merge(brain_all, left_on="id", right_on=0, how="inner")
feature_df_brain
```

```python
signalp = pd.read_csv(data_path + "Features/prediction_results.txt", sep="\t", index_col=False, header=None, skiprows=2,
    names=["Uniprot", "Prediction", "Likelihood-Other", "Likelihood-SP", "CS Position"])

def get_uniprot(string):
    _, uniprot, _ = string.split("|")
    return uniprot

# retrieve Uniprot ID
signalp["Uniprot"] = signalp["Uniprot"].apply(get_uniprot)
signalp_pos = signalp[signalp["Prediction"] == "SP"]
signalp_pos
```

```python
feature_df_brain["Signal peptide"] = np.where(feature_df_brain["id"].isin(signalp_pos["Uniprot"]), 1, 0)
feature_df_brain["Signal peptide"].value_counts()
```

```python
feature_df_brain.drop([0], axis=1, inplace=True)
feature_df_brain
```

```python
brain_elevated = pd.read_csv(data_path + "Brain/HPA_all_brain_elevated_version21.tsv", sep="\t")
brain_elevated.dropna(subset=["Uniprot"], inplace=True)

def keep_first_uniprot(string):
    if "," in string:
        uniprots = string.split(",")
        uniprot1 = uniprots[0]
    else:
        uniprot1 = string
    
    return uniprot1

# if several Uniprots are associated, keep only first
brain_elevated["Uniprot"] = brain_elevated["Uniprot"].apply(keep_first_uniprot)

def get_brain_expression(string):
    # check if expression for multiple tissues is provided
    if ";" in string:
        tissues = string.split(";")
        for t in tissues:
            # keep only information on brain expression
            if "brain" in t:
                brain_string = t 
    else:
        brain_string = string
    
    # extract expression value from string
    _, exp = brain_string.split(" ")
    exp = float(exp)

    return exp

# add column with the mRNA brain expression
brain_elevated["Brain expression"] = brain_elevated["RNA tissue specific nTPM"].apply(get_brain_expression)
brain_elevated
```

```python
feature_df_brain = feature_df_brain.merge(brain_elevated[["Uniprot", "Brain expression"]], left_on="id", right_on="Uniprot", how="inner")
feature_df_brain.drop(["Uniprot"], axis=1, inplace=True)
feature_df_brain
```

```python
feature_df_brain["CSF"] = np.where(feature_df_brain["id"].isin(CSF_pos[0]), 1, -1)
# feature_df_brain.drop(labels=[0], axis=1, inplace=True)
feature_df_brain
```

```python
feature_df_brain["CSF"].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# define explanatory and response variables
X = (feature_df_brain.drop(["CSF", "id", "fasta_sequence"], axis=1))
y = (feature_df_brain["CSF"])

# split 80% training and 20% test 
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=0, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier

# train the model
rf = RandomForestClassifier(random_state=0, n_estimators=1000) 
rf.fit(train_X, train_y)
```

```python
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(rf, test_X, test_y)
plt.show()
```
