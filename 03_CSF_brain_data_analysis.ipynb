## Protein overlap between studies

```python
overlap_array = np.empty((len(study_uniprots), len(study_uniprots)))
overlap_array.fill(np.nan)

for i, study_1 in enumerate(study_uniprots.keys()):
    for j, study_2 in enumerate(study_uniprots.keys()):
        if j > i:
            continue
        intersec = study_uniprots[study_1].intersection(study_uniprots[study_2])
        overlap = len(intersec)/len(study_uniprots[study_1])
        overlap_array[i,j] = overlap
```

```python
sns.set(style=("ticks"), font_scale=1.2)
fig, axs = plt.subplots(1, 1, figsize=(5,5))

# show overlap as value between 0 and 1
sns.heatmap(overlap_array, annot=True, xticklabels=study_names[:6], yticklabels=study_names[:6], vmin=0.5, vmax=1, 
    cmap="YlGnBu", cbar=False)

plt.show() 
fig.savefig(fig_path + "Study_protein_overlap_heatmap.png", bbox_inches="tight", dpi=300, transparent=True)
```

## Bias in mass spectrometry detection

```python
def peptide_category(number):
    """
    Converts integer of peptide number into float peptide number category.
    """
    if number == 0:
        return "0"
    elif number == 1:
        return "1"
    elif number == 2:
        return "2"
    elif number == 3:
        return "3"
    elif number == 4:
        return "4"
    elif number == 5:
        return "5"
    elif number == 6:
        return "6"
    elif number == 7:
        return "7"
    elif number == 8:
        return "8"
    elif number == 9:
        return "9"
    elif 9 < number:
        return "10+"
```

```python
# create list of brain expression of non-CSF, CSF and core CSF in brain elevated HPA proteome
exp_data = [csf_study_num_exp[csf_study_num_exp["#Studies"] == 0]["log2(Brain expression)"],
csf_study_num_exp[csf_study_num_exp["#Studies"] > 0]["log2(Brain expression)"],
csf_study_num_exp[csf_study_num_exp["#Studies"] == 6]["log2(Brain expression)"]]

print("Median log2(brain expression) of")
print("\tnon-CSF proteins: %.2f" % np.median(exp_data[0]))
print("\tCSF proteins: %.2f" % np.median(exp_data[1]))
print("\tcore CSF proteins: %.2f" % np.median(exp_data[2]))
```

```python
# calculate p-values
_, p_nonCSF_CSF = mannwhitneyu(csf_study_num_exp[csf_study_num_exp["#Studies"] == 0]["log2(Brain expression)"],
    csf_study_num_exp[csf_study_num_exp["#Studies"] > 0]["log2(Brain expression)"])
_, p_CSF_coreCSF = mannwhitneyu(csf_study_num_exp[csf_study_num_exp["#Studies"] > 0]["log2(Brain expression)"],
    csf_study_num_exp[csf_study_num_exp["#Studies"] == 6]["log2(Brain expression)"])
_, p_nonCSF_coreCSF = mannwhitneyu(csf_study_num_exp[csf_study_num_exp["#Studies"] == 0]["log2(Brain expression)"],
    csf_study_num_exp[csf_study_num_exp["#Studies"] == 6]["log2(Brain expression)"])

# adjust p-values
adjusted_p_expression = multipletests([p_nonCSF_CSF, p_CSF_coreCSF, p_nonCSF_coreCSF], alpha=0.05, method="fdr_bh")[1]
```

```python
sns.set(style=("ticks"), font_scale=1.2)
fig, ax = plt.subplots(1, 1, figsize=(5,5))

sns.boxplot(data=exp_data, palette=["firebrick", "skyblue", "palegreen"])
ax.set(title="Brain expression across CSF classes", xlabel=None, ylabel="log2(Brain expression)", 
    xticklabels=["non-CSF", "CSF (1+)", "CSF (6+)"], ylim=(-4,30))

# add p-value annotations
# non-CSF vs. CSF
y, h = max(exp_data[0]) + 3, 2
ax.plot([0, 0, 1, 1], [y, y+h, y+h, y], lw=1.5, color="black")
ax.text(0.5, y+2.2, print_p_val(adjusted_p_expression[0]), ha="center", va="bottom", color="black")
# CSF vs. core-CSF
y, h = max(exp_data[1]) + 2, 2
ax.plot([1, 1, 2, 2], [y, y+h, y+h, y], lw=1.5, color="black")
ax.text(1.5, y+2.2, print_p_val(adjusted_p_expression[1]), ha="center", va="bottom", color="black")
# non-CSF vs. core-CSF
y, h = max(exp_data[0]) + 6, 2
ax.plot([0, 0, 2, 2], [y, y+h, y+h, y], lw=1.5, color="black")
ax.text(1, y+2.2, print_p_val(adjusted_p_expression[2]), ha="center", va="bottom", color="black")

plt.show()
fig.savefig(fig_path + "Expression_nonCSF_CSF_coreCSF_brain_elevated.png", bbox_inches="tight", dpi=300, transparent=True)
```

### Check for differences in overlap

```python
# create copy of dictionary with healthy studies to add AD CSF studies to
study_names_AD = ["Higginbotham2020", "Sathe2019", "Bader2020"]
study_uniprots_with_AD = study_uniprots

# create Uniprot lists for each AD CSF data set
for i, col in enumerate(AD_csf.columns[1:4]):
    uniprots = set(AD_csf[AD_csf[col] > 0]["Uniprot"])
    study_uniprots_with_AD[study_names_AD[i]] = uniprots
```

```python
overlap_array = np.zeros((len(study_uniprots_with_AD), len(study_uniprots_with_AD)))

for i, study_1 in enumerate(study_uniprots_with_AD.keys()):
    for j, study_2 in enumerate(study_uniprots_with_AD.keys()):
        intersec = study_uniprots_with_AD[study_1].intersection(study_uniprots_with_AD[study_2])
        overlap = len(intersec)/len(study_uniprots_with_AD[study_1])
        overlap_array[i,j] = overlap
```

```python
sns.set(style=("ticks"), font_scale=1.2)
fig, axs = plt.subplots(1, 1, figsize=(8,8))

# show overlap as value between 0 and 1
sns.heatmap(overlap_array, annot=True, xticklabels=study_names[:6]+study_names_AD, 
    yticklabels=study_names[:6]+study_names_AD, vmin=0, vmax=1, cmap="YlGnBu", cbar=True)

plt.show() 
```

```python
sns.set(style=("ticks"), font_scale=1.2)
fig, ax = plt.subplots(1, 1, figsize=(5,5))

overlap_healthy_dict = {}
overlap_AD_dict = {}

for i, study in enumerate(study_uniprots_with_AD.keys()):

    overlap_healthy = []
    overlap_AD = []
    
    # remove current study from comparison
    studies_compare = list(study_uniprots_with_AD.keys())
    studies_compare.remove(study)
    
    assert len(studies_compare) == 8
    
    for study_2 in studies_compare:
        
        # get set of overlapping proteins
        intersection = study_uniprots_with_AD[study].intersection(study_uniprots_with_AD[study_2])
        # overlap as fraction of dataset that we compare with
        overlap = len(intersection)/len(study_uniprots_with_AD[study_2])
        
        # add overlaps to dictionaries and add point to scatter plot
        if study_2 in ["Macron2018A", "Macron2020", "Zhang2015", "Guldbrandsen2014", "Macron2018B", "Schutzer2010"]:
            overlap_healthy.append(overlap)
            scatter1 = plt.scatter(x=overlap, y=study, color="skyblue", label="Healthy CSF")
        if study_2 in ["Higginbotham2020", "Sathe2019", "Bader2020"]:
            overlap_AD.append(overlap)
            scatter2 = plt.scatter(x=overlap, y=study, color="navy", label="AD CSF")    

plt.legend(handles=[scatter1, scatter2], loc="lower left")
ax.set(title="Overlap across healthy and AD CSF studies", xlabel="Overlap of identified unique proteins", xlim=(0,1))
    
plt.show()
fig.savefig(fig_path + "Scatter_overlap_healthy_AD_CSF_proteins.png", bbox_inches="tight", dpi=300, transparent=True)
```
