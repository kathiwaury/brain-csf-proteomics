```python
# import libraries 
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns

from matplotlib import patches
from matplotlib_venn import venn2
from pathlib import Path
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import multipletests

# ignore pandas copy warning
import warnings
from pandas.core.common import SettingWithCopyWarning
warnings.simplefilter(action="ignore", category=SettingWithCopyWarning)
```

# Import results
https://prosite.expasy.org/scanprosite/scanprosite_doc.html#of_graphical


```python
data_path = str(Path(os.getcwd()).parents[0]) + "/Motif_discovery/"
fig_path = str(Path(os.getcwd()).parents[0]) + "/Figures/"
```

## PROSITE full scan results

```python
prosite_CSF = pd.read_csv(data_path + "PROSITE_CSF.txt", sep="\t", index_col=False, header=None, 
    names=["Uniprot", "Start", "End", "PROSITE motif", "Profile score", "Confidence (miniprofile)", "Confidence", 
           "Matched region"], skiprows=7)
prosite_CSF = prosite_CSF.loc[:, ["Uniprot", "PROSITE motif"]]
```

```python
prosite_non_CSF = pd.read_csv(data_path + "PROSITE_non_CSF.txt", sep="\t", index_col=False, header=None, 
    names=["Uniprot", "Start", "End", "PROSITE motif", "Profile score", "Confidence (miniprofile)", "Confidence", 
           "Matched region"], skiprows=7)
prosite_non_CSF = prosite_non_CSF.loc[:, ["Uniprot", "PROSITE motif"]]
```

```python
def get_uniprot(string):
    _, uniprot, _ = string.split("|")
    return uniprot
```

```python
# retrieve Uniprot ID
prosite_CSF["Uniprot"] = prosite_CSF["Uniprot"].apply(get_uniprot)
# retrieve PROSITE ID and pattern name
prosite_CSF[["PROSITE motif", "Pattern name"]] = prosite_CSF["PROSITE motif"].str.split("|", expand=True)
# drop duplicate pattern in same protein
prosite_CSF.drop_duplicates(subset = ["Uniprot", "PROSITE motif"], inplace=True)
prosite_CSF[:5]
```

```python
# retrieve Uniprot ID
prosite_non_CSF["Uniprot"] = prosite_non_CSF["Uniprot"].apply(get_uniprot)
# retrieve PROSITE ID and pattern name
prosite_non_CSF[["PROSITE motif", "Pattern name"]] = prosite_non_CSF["PROSITE motif"].str.split("|", expand=True)
# drop duplicate pattern in same protein
prosite_non_CSF.drop_duplicates(subset = ["Uniprot", "PROSITE motif"], inplace=True)
prosite_non_CSF[:5]
```

```python
pattern_count_CSF = prosite_CSF["Pattern name"].value_counts()
pattern_count_non_CSF = prosite_non_CSF["Pattern name"].value_counts()
```

```python
# overlap of PROSITE patterns
pos_set = set(prosite_CSF["PROSITE motif"])
neg_set = set(prosite_non_CSF["PROSITE motif"])

venn2([pos_set, neg_set], set_labels=("CSF", "non-CSF"))
plt.show()
```

```python
# merge positive and negative CSF associated patterns
pattern_count = pd.concat([pattern_count_CSF, pattern_count_non_CSF], axis=1)
pattern_count.fillna(0, inplace=True)
pattern_count.columns = ["CSF", "non-CSF"]
```

```python
def fishers_test(df):

    A = int(df.values[0:1]) # CSF  & pattern yes
    B = int(num_CSF - df.values[0:1]) # CSF  & pattern no
    C = int(df.values[1:2]) # non-CSF & pattern yes
    D = int(num_non_CSF - df.values[1:2]) # non-CSF & pattern no

    # create 2x2 contingency table of counts
    table = np.array([[A, C], [B, D]])
    table
    _, p = fisher_exact(table, alternative="two-sided") # function return odds ratio and p-value
    return p
```

```python
# relative abundance of patterns in CSF positive and negative protein set
num_CSF = 953
num_non_CSF = 1593

pattern_count["CSF fraction"] = pattern_count["CSF"].apply(lambda x: x/num_CSF)
pattern_count["non-CSF fraction"] = pattern_count["non-CSF"].apply(lambda x: x/num_non_CSF)
pattern_count["Difference"] =  pattern_count["CSF"] - pattern_count["non-CSF"]
# calculate p-value (Fisher's exact test, chi square)
pattern_count["P-val"] = pattern_count.apply(fishers_test, axis=1)
# calculate adjusted p-value 
# rejected, p_adj = fdrcorrection(pattern_count["P-val"])
adjusted_p = multipletests(pattern_count["P-val"], alpha=0.05, method="fdr_bh")[1]
pattern_count["P-val adjusted"] = adjusted_p
pattern_count.sort_values("P-val adjusted", inplace=True)
```

```python
# look only at domain with significant difference between groups
pattern_count_sig = pattern_count[pattern_count["P-val adjusted"] < 0.05]
pattern_count_sig.sort_values("Difference", ascending=False, inplace=True)
```

```python
# create melted dataframe
pattern_count_sig_less = pattern_count_sig[["CSF fraction", "non-CSF fraction"]]
pattern_count_sig_less.reset_index(inplace=True)
pattern_count_melted = pattern_count_sig_less.melt(id_vars=["index"], value_vars=["CSF fraction", "non-CSF fraction"], 
    var_name="CSF", value_name="Fraction")
pattern_count_melted
```

```python
sns.set(style=("ticks"), font_scale=1)
fig, ax = plt.subplots(1, 1, figsize=(12,8))

palette = {"CSF fraction": "skyblue", "non-CSF fraction": "firebrick"}

bar = sns.barplot(x="index", y="Fraction", data=pattern_count_melted, hue="CSF", palette=palette)
bar.legend(title="")
bar.set(xlabel=None, ylabel="Fraction", title="Percentage of brain proteins associated with PROSITE domains and patterns")

# create legend
P = patches.Patch(color="skyblue", label="CSF")
N = patches.Patch(color="firebrick", label="non-CSF")
plt.legend(title="Class", loc="upper center", handles=[P, N]);

fig.autofmt_xdate(rotation=90)

plt.show()
fig.savefig(fig_path + "PROSITE_domains_brain_proteins.png", bbox_inches="tight", dpi=300, transparent=True)
```
